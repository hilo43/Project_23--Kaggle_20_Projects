# Feature Engineering

```python
df["name"].value_counts()
```

```python
df['Age'] = 2022 - df['year']
df.drop('year',axis=1,inplace = True)
```

```python
df["fuel"].value_counts()
```

```python
df.groupby("fuel")["selling_price"].mean().sort_values()
```

```python
df.replace({'fuel':{'LPG':1,'CNG':2 , "Electric":3 , "Petrol":4 , "Diesel":5}  },inplace=True)
```

```python
df["seller_type"].value_counts()
```

```python
df.replace({'seller_type':{'Dealer':1,'Individual':2,'Trustmark Dealer':3}},inplace=True)
```

```python
df.info()
```

```python
df["transmission"].value_counts()
```

```python
df.groupby("transmission")["selling_price"].mean().sort_values()
```

```python
df.replace({'transmission':{'Manual':0,'Automatic':1}},inplace=True)
```

```python
df["owner"].value_counts()
```

```python
df.groupby("owner")["selling_price"].mean().sort_values()
```

```python
df.replace({'owner':{'First Owner':4,'Second Owner':3,'Third Owner':2,'Fourth & Above Owner':1,'Test Drive Car':5}},inplace=True)
```

```python
df = df.drop("name" , axis=1)
```

```python
df.head()
```

```python
abs(df.corr()["selling_price"]).sort_values(ascending = False) 
```
